```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://${DB_USER}:${DB_PASS}@localhost:5432/ny_taxi')
```

```python
df_zones = pd.read_csv('data/taxi_zone_lookup.csv')
```

```python
# Get schemas
print(pd.io.sql.get_schema(df_zones, name='green_taxi_data', con=engine))
```

```python
# Inject the data
df_zones.to_sql(name='zones', con=engine, if_exists='replace')
```
